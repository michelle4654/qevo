```python
import numpy as np
import itertools


lF = np.log(.97)
lqs = [np.log(.01)]*3
probs = np.empty((4,)*3)
yo = zip(itertools.product(range(4), repeat = 3), itertools.product((lF, *lqs), repeat = 3))
for i, p in yo: probs[i] = np.exp(sum(p))
probs
```

```python
qevo.depolarize2(probs,(0,1),.97)
```

```python
qevo.depolarize2(probs,(1,2),.97)
```
